```python
# Import full dataset

df_saved = pd.read_csv('df_pricesandfeatures_for_EDA')
df = df_saved.iloc[:, 1:] # get rid of index column

# deal with numbers scraped as strings
df['GDPpercap_nominal'] = [str(value).replace(',', '') for value in df.GDPpercap_nominal]
df['GDPpercap_nominal'] = [float(value) for value in df.GDPpercap_nominal]
df['GDPpercap_PPP'] = [str(value).replace(',', '') for value in df.GDPpercap_PPP]
df['GDPpercap_PPP'] = [float(value) for value in df.GDPpercap_PPP]

df.drop(['URL', 'admin_name'], axis=1, inplace=True) # don't really need

# df.head(20)
```

```python
# Column labels and subsets

# ID data
colnames_id = ['City', 'Country']

# Price data
colnames_COL = ['COL (Family of 4)', 'COL (Single of 1)']

colnames_prices = ['rent_85m2_expens', 'rent_85m2_normal', 'rent_45m2_expens',
                   'rent_45m2_normal', 'eatout_lunch', 'eatout_dinner',
                   'taxi_5mi', 'gas_liter', 'pubtrans_monthly', 'internet_monthly',
                   'TV_40in', 'cappuccino', 'mobile_wifi_128gb']

# Feature data
colnames_country_float = ['GDPpercap_nominal', 'GDPpercap_PPP', 'HDI']

colnames_country_categ = ['Region', 'Subregion']

colnames_city_float = ['population', 'lat', 'lng']

colnames_city_categ = ['Climate', 'capital']

# Contribution info
colnames_contrib = ['No. Prices', 'No. Contributors']

# Price features
# Screened to limit collinearity (e.g. four terms for 'rent')

colnames_price_features = ['rent_45m2_normal', 'eatout_dinner', 'taxi_5mi', 
                   'pubtrans_monthly', 'internet_monthly', 'cappuccino']

# Aggregate lists
colnames_features = colnames_country_categ + colnames_city_categ + colnames_city_float + colnames_country_float
colnames_numeric = colnames_COL + colnames_country_float + colnames_city_float + colnames_prices
colnames_features_num = colnames_country_float + colnames_city_float
colnames_features_num_plusprices = colnames_country_float + colnames_city_float + colnames_prices

# Preferred order
df = df[colnames_id + colnames_COL + colnames_features + colnames_prices] # or price features?
```

### Fill NaNs

```python
# df.info()   # check dtypes, non-null counts
```

```python
# Price data missing from USD countries due to scraping issue
# Come back if time...

list_usd_countries = ['United States', 'Ecuador', 'Puerto Rico']

df_no_usd = df[(df['Country'] != 'United States') &
             (df['Country'] != 'Ecuador') &
             (df['Country'] != 'Puerto Rico')]

df = df_no_usd.reset_index(drop=True)
df # shape: (1064, 27)
```

```python
# Replace NaN values of categorical variables
# Come back if time... (split into trainval/test BEFORE dummy/poly creation

# 'capital' = NaN denotes cities that are not national, regional capitals
df['capital'].fillna(value='blank', inplace=True)

# Code from class notebook
# df[column_name].fillna(value=df[column_name].mean(), inplace=True)
```

```python
# Remove NaNs of continuous variables
# Replace with mean of - in order of priority - city's country, subregion, region

list_of_countries = list(df['Country'].unique())
                 
for index, row in df.iterrows():
    
    country = row.Country
    subregion = row.Subregion
    region = row.Region
    
    if country in list_of_countries:  # needed to skip United States, Ecuador, Puerto Rico
        
        # create dataframes for city's country, subregion
        
        df_same_country = df[(df['Country'] == country)]
        df_same_subregion = df[(df['Subregion'] == subregion)]
        df_same_region = df[(df['Region'] == region)]
        
        # check each column (feature)         
        
        for column in colnames_numeric:              # colnames_numeric or colnames_prices
            
            country_mean = np.mean(df_same_country[column])
            subregion_mean = np.mean(df_same_subregion[column])
            region_mean = np.mean(df_same_region[column])
            
            if np.isnan(country_mean):       # no data on this feature/country, take subregion mean
                df.loc[index, column] = subregion_mean
            elif np.isnan(subregion_mean):   # no data on this feature/subregion, take region mean
                df.loc[index, column] = region_mean
            else:                            # 1+ data on this feature, take country mean
                df.loc[index, column] = country_mean
    else:
        pass

# Remaining NaNs present in rows without Wikipedia data (e.g. HDI)...
df = df.dropna(subset=['HDI']).reset_index(drop=True)

# Test for United States (omitted first time)
# df_test = df2[(df2['Country'] != 'United States')]

df.shape # (1041, 27)
```
