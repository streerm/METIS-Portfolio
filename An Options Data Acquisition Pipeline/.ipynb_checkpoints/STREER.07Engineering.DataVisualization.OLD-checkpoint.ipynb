### 4. Generate Dataframe [OLD]
Rows = Strike price  
Cols = Expiry date
Keep for now: useful to have all strikes & all dates in one place.

```python
df = pd.DataFrame(test_json)#.iloc[50:100, :]

# First, get the rows (strike prices) sorted ascending.
# They are type 'str' in the data, and must be converted to 'float' to sort, then switched back to string
idx_new = [float(num) for num in list(df.index)]
df.index = pd.Index(idx_new)
df2 = df.sort_index()
idx_new = [str(num) for num in list(df2.index)]
df2.index = pd.Index(idx_new)
df2
```

```python
# Next, create an empty dataframe with same row/col indices, to which data will be sent:

df_new = df2.copy()

for col in df_new.columns:
    df_new[col].values[:] = np.nan
df_new
```

```python
# Get list of expiry dates (top level in document)
dates = list(test_json.keys())
print(f'{len(dates)} columns')
print(dates)
```

```python
# Get list of all possible strikes (next level)
strikes = []
for date in dates:
    db = test_json[date]
    strike = [float(key) for key in db.keys()]
    strikes += strike
    
strikes = [str(strike) for strike in list(np.unique(strikes))]  # type conversion needed bc np.unique() sorts alphabetically (e.g. 60>500)

print(f'{len(strikes)} rows')
print(strikes)
```

```python
for date in dates:
    for strike in strikes:
        if df.loc[strike, date]: # if such a strike/expiry option combi existed in the raw data, and 
            if df.loc[strike, date] is not np.nan:   # if that combi contained data, take the midpoint of the bid/ask spread
                df_new.loc[strike, date] = (df.loc[strike, date][0]['ask'] + df.loc[strike, date][0]['bid']) / 2
                
df_new
```

### 4. Generate Dataframe

```python
# SUPERIOR approach

def prep_df_from_json(json):
    df = pd.DataFrame(json)
    vals = list(df.values.flatten())
    
    df_new = []
    for entry in vals:
        if entry is not np.nan:
            df_new.append(entry[0])
    
    print(f'Total of {len(df_new)} contracts')
        
    df = pd.DataFrame(tuple(df_new))
    
    return df

df_c = prep_df_from_json(test_json_c)
df_p = prep_df_from_json(test_json_p)
df = pd.concat([df_c, df_p], axis=0, ignore_index=True)
df.to_pickle("df_test.pkl")
df.to_csv('df_test.csv')
```

```python
# Visualization #1: 3D scatter plot

x = df_c['strikePrice']
y = df_c['expirationDate']
z = df_c['mark']

fig = plt.figure(figsize=(10,100))
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z)
ax.set_box_aspect((1, 3, 1))

ax.set_xlabel('Strike Price ($)')
ax.set_ylabel('Expiration Date (epoch)')
ax.set_zlabel('Option Price ($)')

#ax.set_xlim(left=200, right=400)
ax.set_ylim(bottom=1639774800000, top=1645496000000)
#ax.set_zlim(bottom=0, top=1000)

ax.view_init(20, -60)
plt.show()
```

```python
"""
Example of a line chart of time-series simulation in Matplotlib
"""

import streamlit as st
import numpy as np
import matplotlib.pyplot as plt

st.title("Today's NVDA Options Chain")
st.write("'We have to realize that computers are simulators and then figure out what to simulate.' \n\n — Alan Kay")

trend = st.slider('Trend',  min_value=0.001, max_value=0.10, step=0.01)
#date = st.slider('Date', min_value='')
noise = st.slider('Noise',min_value=0.01,  max_value=0.10, step=0.01)
st.write(f"Trend = {trend} \n\n Noise = {noise}")

intial_value = 1
n_series = 10 
time_series = np.cumprod(intial_value + np.random.normal(trend, noise, (100, n_series)), 
                         axis=0)

st.line_chart(time_series)

fig, ax = plt.subplots()
for ts in time_series.T:
    ax.plot(ts)

st.pyplot(fig)

# Notes:
# - Switch from function to procedural
# - Lag in rendering
# - When you deploy you have to add external dependency file (not covered)
```
