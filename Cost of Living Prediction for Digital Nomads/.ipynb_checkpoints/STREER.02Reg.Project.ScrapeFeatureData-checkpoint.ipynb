### Get Human Development Index (HDI) by country
Source: [Wikipedia](https://en.wikipedia.org/wiki/List_of_countries_by_Human_Development_Index)

```python
url = 'https://en.wikipedia.org/wiki/List_of_countries_by_Human_Development_Index'  # generate page version with USD
    
response = requests.get(url)                       
page = response.text
soup_wiki = BeautifulSoup(page, "lxml")
    
ripped = []

# start at the table
chunk = soup_wiki.find_all('tr')[2]        
country_raw = chunk.find('a')

for i in range(191):
    country = country_raw.text
    HDI = country_raw.findNext().text
    
    ripped.append((country, HDI))
    
    country_raw = country_raw.findNext('a')

    
del ripped[116]
del ripped[117]
    
ripped
```

```python
df = pd.DataFrame(ripped, columns = ['Country', 'HDI'])
df.info()
```

### Get GDP per capita (nominal/PPP) and regions by country

```python
url1 = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)_per_capita'

response = requests.get(url1)                       
page = response.text
soup_wiki = BeautifulSoup(page, "lxml")
    
ripped1 = []

# start at the table
chunk = soup_wiki.find_all('tr')[3]        
country_raw = chunk.find('a')

for i in range(39):     # URLs in 39th entry - Taiwan - create error

    country = country_raw.text
    subregion = country_raw.findNext('td').text
    region = country_raw.findNext('td').findNext('td').text
    GDPpercap_nominal = country_raw.findNext('td').findNext('td').findNext('td').text

    ripped1.append((country[:-2], subregion, region, GDPpercap_nominal))
    
    country_raw = country_raw.findNext('a').findNext('a').findNext('a')

ripped2 = []

chunk2 = soup_wiki.find_all('tr')[42]        # Bahamas
country_raw = chunk2.find('a')

for i in range(174):

    country = country_raw.text
    subregion = country_raw.findNext('td').text
    region = country_raw.findNext('td').findNext('td').text
    GDPpercap_nominal = country_raw.findNext('td').findNext('td').findNext('td').text

    ripped2.append((country[:-2], subregion, region, GDPpercap_nominal))
    
    country_raw = country_raw.findNext('a').findNext('a').findNext('a')

ripped = ripped1 + ripped2
ripped
```

```python
df = pd.DataFrame(ripped, columns = ['Country', 'Subregion', 'Region', 'GDPpercap_nominal'])
df.info()
```

```python
df.to_pickle('df_countryGDP_nom')
df.to_csv('df_countryGDP_nom', index=False)
df.head(10)
```

### Climate classification
Source: [University of Vienna](http://koeppen-geiger.vu-wien.ac.at/present.htm)

```python
climate_ref = pd.read_csv('Koeppen-Geiger-ASCII.txt')
climate_ref.info()
```

```python
list_Lat = []
list_Lon = []
list_Cls = []

for index, row in climate_ref.iterrows():

    full_str = climate_ref.iloc[index, 0].strip()
    list_three = [entry for entry in full_str.split(' ') if entry != '']
    
    list_Lat.append(list_three[0])
    list_Lon.append(list_three[1])
    list_Cls.append(list_three[2])
    
list_tuples = list(zip(list_Lat, list_Lon, list_Cls))

df_climate = pd.DataFrame(list_tuples,
                       columns = ['Lat', 'Lon', 'Cls'])

df_climate.to_csv('df_climate_reference', index=False)

df_climate
```

```python
from math import sqrt

# for index, row in df_reliable2.iterrows():
    
city_lat = float(df_reliable2.iloc[0, :]['lat'])
city_lng = float(df_reliable2.iloc[0, :]['lng'])

# city_lat, city_lng = row['lat'], row['lng']
city_lat, city_lng
```

```python
min_dist = 1000 # arbitrary starting point

for index2, row2 in df_climate.iterrows():
    clim_lat, clim_lng = float(row2['Lat']), float(row2['Lon'])

    test_dist = sqrt((city_lat - clim_lat)**2 + (city_lng - clim_lng)**2)
    if test_dist < min_dist:
        min_dist = test_dist
        climate = row2['Cls']

climate
# df_reliable.iloc[index]
```

```python
city_climates = []

for index, row in df_reliable2.iterrows():
    city_lat, city_lng = float(row['lat']), float(row['lng'])
    
    min_dist = 1000 # arbitrary starting point

    for index2, row2 in df_climate.iterrows():
        clim_lat, clim_lng = float(row2['Lat']), float(row2['Lon'])
        
        test_dist = sqrt((city_lat - clim_lat)**2 + (city_lng - clim_lng)**2)
        
        if test_dist < min_dist:
            min_dist = test_dist
            climate = row2['Cls']
    
    city_climates.append(climate)
    print(f'City: {row.City}, Climate: {climate}')
```
